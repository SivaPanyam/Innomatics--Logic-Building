```python
# User Login Check
# Problem: Verify if the provided username and password match the expected credentials

# Define the expected credentials
expected_username = "admin"
expected_password = "1234"

# Get user input (In real scenarios, these would come from user input)
username = "admin"
password = "1234"

# Check if both username and password match
if username == expected_username and password == expected_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# Pass / Fail Analyzer
# Problem: Analyze student results and count pass/fail students

# Given student marks
marks = [45, 78, 90, 33, 60]

# Define passing criteria
passing_marks = 50

# Initialize counters
pass_count = 0
fail_count = 0

# Analyze each student's marks
for mark in marks:
    if mark >= passing_marks:
        pass_count += 1
    else:
        fail_count += 1

# Print the results
print("=" * 40)
print("Student Result Analysis")
print("=" * 40)
print(f"Total Students: {len(marks)}")
print(f"Pass Students: {pass_count}")
print(f"Fail Students: {fail_count}")
print("=" * 40)
```

```python
# Message Length Analyzer
# Analyze message sizes and flag long messages

# Given messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Threshold for long messages
threshold = 10

# Analyze each message
print("Message Length Analysis:")
print("-" * 50)

for i, message in enumerate(messages, 1):
    length = len(message)
    flag = " [LONG MESSAGE]" if length > threshold else ""
    print(f"Message {i}: '{message}' - Length: {length}{flag}")

print("-" * 50)
```

```python
# Error Message Detector
# Detect and count error messages from system logs

# System logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Count ERROR entries
error_count = logs.count("ERROR")

# Print the total error count
print(f"Total ERROR count: {error_count}")
```
