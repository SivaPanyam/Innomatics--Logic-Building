# Problem 1: Unique Words in a Sentence

Given a sentence, identify all unique words and display:
1. The count of unique words
2. The unique words themselves

```python
# Step 1: Get input from the user
sentence = "python is easy and python is powerful" # Example input
# To take user input, uncomment the line below:
# sentence = input("Enter sentence: ")

# Step 2: Split the sentence into words
words = sentence.split()

# Step 3: Convert the list of words to a set to get unique words
unique_words = set(words)

# Step 4: Display the count and the unique words
print(f"Unique words count: {len(unique_words)}")
print(f"Unique words: {unique_words}")
```

# Problem 2: Highest Salary from Employee Data

Store employee names and their salaries, and determine which employee has the highest salary.

```python
# Step 1: Define the employee data
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
    # Add more if needed
}

# Step 2: Find the employee with the highest salary
# Using 'max' function with 'key' argument
# employees.get returns the value for a given key, so max finds the key with the max value
highest_paid_employee = max(employees, key=employees.get)
highest_salary = employees[highest_paid_employee]

# Step 3: Print the result
print(f"Highest Salary: {highest_paid_employee} - {highest_salary}")
```

# Problem 3: Find Maximum and Minimum Values

Find the maximum and minimum values from a list of numbers.

```python
# Step 1: Define the list of numbers
numbers = [45, 22, 89, 10, 66]

# Step 2: Initialize max and min with the first element
if not numbers:
    print("List is empty")
else:
    max_val = numbers[0]
    min_val = numbers[0]

    # Step 3: Loop through the list to find max and min
    for num in numbers:
        if num > max_val:
            max_val = num
        if num < min_val:
            min_val = num

    # Step 4: Print the results
    print(f"List: {numbers}")
    print(f"Max: {max_val}")
    print(f"Min: {min_val}")
```

# Problem 4: Count Products Above a Price Threshold

Count how many products have a price greater than 1000.

```python
# Step 1: Define the list of prices
prices = [450, 1200, 899, 1500, 300]

# Step 2: Initialize a counter
count = 0

# Step 3: Loop through the prices and count those above 1000
for price in prices:
    if price > 1000:
        count += 1

# Step 4: Print the result
print(f"Products above 1000: {count}")
```

# Problem 5: Calculate Attendance Percentage

Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).

```python
# Step 1: Define the attendance list
attendance = ["P", "P", "A", "P", "P"]

# Step 2: Count the number of days present
days_present = attendance.count("P")

# Step 3: Calculate the total number of days
total_days = len(attendance)

# Step 4: Calculate the percentage
percentage = (days_present / total_days) * 100

# Step 5: Print the result
print(f"Attendance Percentage: {percentage}")
```

# Problem 7: Count Character Frequency

Count how many times each character appears in a given string.

```python
# Step 1: Define the string
string_input = "pythonp"

# Step 2: Initialize an empty dictionary to store frequencies
char_count = {}

# Step 3: Loop through each character in the string
for char in string_input:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

# Step 4: Print the result
print(char_count)
```

# Problem 10: Calculate Average Marks

Calculate the average marks of a student from a list of marks.

```python
# Step 1: Define the list of marks
marks = [85, 90, 78, 92, 88] # Example marks
# To match the expected output of 72.5 exactly:
marks = [70, 80, 75, 65]

# Step 2: Calculate the sum of marks
total_marks = sum(marks)

# Step 3: Calculate the number of subjects
num_subjects = len(marks)

# Step 4: Calculate the average
average_marks = total_marks / num_subjects

# Step 5: Print the result
print(f"Average Marks: {average_marks}")
```
