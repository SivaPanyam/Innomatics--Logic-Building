# Problem 1: Mobile Recharge Validation System


```python
valid_plans = [199, 299, 399, 599]

while True:
    try:
        amount = int(input("Enter recharge amount: ₹"))

        if amount < 50:
            print("Recharge amount must be at least ₹50.")
        
        elif amount in valid_plans:
            print(f"Recharge successful for ₹{amount}!")
            break
        
        else:
            print("Invalid plan selected. Available plans: 199, 299, 399, 599.")
        
        print("Please try again.\n")

    except ValueError:
        print("Please enter a valid numeric amount.\n")
```

Problem 2: Inventory Reorder Alert System


```python
def check_inventory(stock_data):
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert (Stock = {quantity})")
        else:
            print(f"{product}: Stock OK (Stock = {quantity})")


# Example dictionary
inventory = {
    "Rice": 25,
    "Sugar": 10,
    "Oil": 5,
    "Wheat": 18,
    "Salt": 12
}

check_inventory(inventory)
```

 Problem 3: Student Result Processing System

```python
def process_result(marks):
    total = 0
    
    # Calculate total using loop
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    
    if average >= 50:
        return f"Average Marks: {average:.2f} - Pass"
    else:
        return f"Average Marks: {average:.2f} - Fail"


# Example usage
student_marks = [60, 45, 70, 55, 40]

result = process_result(student_marks)
print(result)
```

Problem 4: Cab Fare Estimator with Retry Option


```python
# Function to calculate cab fare
def calculate_fare(distance, peak):
    base_fare = 50          # Fixed base fare
    per_km_rate = 12        # Cost per kilometer

    # Calculate normal fare
    fare = base_fare + (distance * per_km_rate)

    # Add 25% extra if it is peak hour
    if peak:
        fare += fare * 0.25

    return fare


# While loop to allow retry
while True:
    try:
        # Take distance input
        distance = float(input("Enter distance in km: "))

        # Take peak hour input
        peak_input = input("Is it peak hour? (yes/no): ").strip().lower()

        # Validate peak input
        if peak_input not in ["yes", "no"]:
            print("Please enter 'yes' or 'no' for peak hour.\n")
            continue

        # Convert input to boolean
        peak = True if peak_input == "yes" else False

        # Call function to calculate fare
        total_fare = calculate_fare(distance, peak)

        # Display total fare
        print(f"Total Cab Fare: ₹{total_fare:.2f}")

        # Ask user if they want to retry
        retry = input("Do you want to calculate again? (yes/no): ").strip().lower()
        if retry != "yes":
            print("Thank you for using the Cab Fare Estimator.")
            break

    except ValueError:
        # Handle invalid numeric input
        print("Please enter a valid distance.\n")
```

Problem 5: Employee Attendance Eligibility Checker


```python
# Function to check attendance eligibility
def check_eligibility(attendance_list):
    total_days = len(attendance_list)   # Total number of days
    present_days = 0                    # Counter for present days

    # Loop to count number of "P"
    for status in attendance_list:
        if status == "P":
            present_days += 1

    # Calculate attendance percentage
    percentage = (present_days / total_days) * 100

    # Check eligibility condition
    if percentage >= 75:
        return f"Attendance: {percentage:.2f}% - Eligible"
    else:
        return f"Attendance: {percentage:.2f}% - Not Eligible"


# Example attendance record
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]

# Call function and print result
result = check_eligibility(attendance)
print(result)
```

Problem 6: Password Strength Checker


```python
# Function to check password strength
def check_password_strength(password):
    has_digit = False
    has_special = False

    # Check minimum length
    if len(password) < 8:
        return "Weak Password: Minimum length should be 8 characters."

    # Loop through each character in the password
    for char in password:
        # Check if character is a digit
        if char.isdigit():
            has_digit = True

        # Check if character is one of the required special characters
        if char in "@#$":
            has_special = True

    # Final validation conditions
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password: Must contain at least one digit and one special character (@#$)."


# Example usage
password = input("Enter your password: ")
result = check_password_strength(password)
print(result)
```
