```python
# Pass / Fail Analyzer
# Problem: Analyze student results and count pass/fail students

# Given student marks
marks = [45, 78, 90, 33, 60]

# Define passing criteria
passing_marks = 50

# Initialize counters
pass_count = 0
fail_count = 0

# Analyze each student's marks
for mark in marks:
    if mark >= passing_marks:
        pass_count += 1
    else:
        fail_count += 1

# Print the results
print("=" * 40)
print("Student Result Analysis")
print("=" * 40)
print(f"Total Students: {len(marks)}")
print(f"Pass Students: {pass_count}")
print(f"Fail Students: {fail_count}")
print("=" * 40)
```
