Problem 1: Smart Light Controller

Problem Statement

Create a class to control a smart light that can be turned ON or OFF and display its status.


```python
# Class to represent a Smart Light
class SmartLight:
    
    # Constructor to initialize the light name and default status
    def __init__(self, light_name):
        self.light_name = light_name   # Stores the name of the light
        self.status = False            # Light is OFF by default

    # Method to turn the light ON
    def turn_on(self):
        self.status = True

    # Method to turn the light OFF
    def turn_off(self):
        self.status = False

    # Method to display the current status of the light
    def display_status(self):
        if self.status:
            print(self.light_name + " is ON")
        else:
            print(self.light_name + " is OFF")


# Taking input from the user
light_name = input("Light Name: ")
action = input("Action: ")

# Creating an object of the SmartLight class
light = SmartLight(light_name)

# Checking the action entered by the user
if action.upper() == "ON":
    light.turn_on()     # Turn the light ON
else:
    light.turn_off()    # Turn the light OFF

# Display the current status of the light
light.display_status()
```

Problem 2: Employee ID Card System

Problem Statement

Create a class that stores employee name, ID, and department and displays ID card details.


```python
# Class to represent an Employee ID Card
class Employee:
    
    # Constructor to initialize employee details
    def __init__(self, name, emp_id, department):
        self.name = name              # Stores employee name
        self.emp_id = emp_id          # Stores employee ID
        self.department = department  # Stores employee department

    # Method to display the employee ID card details
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Taking input from the user
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")

# Creating an object of the Employee class
employee = Employee(name, emp_id, department)

# Displaying the employee ID card
employee.display_id_card()
```

Problem 3: Mobile Contact Record

Problem Statement

Create a class to store contact name and phone number and display contact information.


```python
# Class to represent a Mobile Contact
class Contact:

    # Constructor to initialize contact name and phone number
    def __init__(self, name, phone):
        self.name = name      # Stores the contact name
        self.phone = phone    # Stores the contact phone number

    # Method to display the saved contact details
    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Taking input from the user
name = input("Contact Name: ")
phone = input("Phone Number: ")

# Creating an object of the Contact class
contact = Contact(name, phone)

# Displaying the saved contact information
contact.display_contact()
```

Problem 4: Product Price Tag Generator

Problem Statement

Create a class that stores product name and price and prints a formatted price tag.


```python
# Class to represent a Product
class Product:

    # Constructor to initialize product name and price
    def __init__(self, name, price):
        self.name = name     # Stores the product name
        self.price = price   # Stores the product price

    # Method to display the formatted price tag
    def display_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


# Taking input from the user
name = input("Product Name: ")
price = input("Price: ")

# Creating an object of the Product class
product = Product(name, price)

# Displaying the price tag
product.display_price_tag()
```

Problem 5: Movie Rating Display System

Problem Statement

Create a class that stores movie name and rating and displays movie details.


```python
# Class to represent a Movie
class Movie:

    # Constructor to initialize movie name and rating
    def __init__(self, name, rating):
        self.name = name      # Stores the movie name
        self.rating = rating  # Stores the movie rating

    # Method to display the movie details
    def display_movie(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


# Taking input from the user
movie_name = input("Movie Name: ")
rating = input("Rating: ")

# Creating an object of the Movie class
movie = Movie(movie_name, rating)

# Displaying the movie details
movie.display_movie()
```

Problem 6: Delivery Address Manager

Problem Statement

Create a class that stores customer name and delivery address and prints delivery details.


```python
# Class to represent Delivery Details
class Delivery:

    # Constructor to initialize customer name and address
    def __init__(self, name, address):
        self.name = name        # Stores the customer name
        self.address = address  # Stores the delivery address

    # Method to display delivery details
    def display_details(self):
        print("Delivery Details")
        print("Customer:", self.name)
        print("Address:", self.address)


# Taking input from the user
customer_name = input("Customer Name: ")
address = input("Address: ")

# Creating an object of the Delivery class
delivery = Delivery(customer_name, address)

# Displaying the delivery information
delivery.display_details()
```
